```python
# Exercise 2
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result
x=print(func2(2,4))
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
x=print(func4([1, 2, 3, 4], [2, 4, 6, 8]))
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
# sample function gives the common element present in 2 diffent list
import time
start = time.time()
x=print(func4([1, 2, 3, 4], [2, 4, 6, 8]))
end=time.time()
print(end-start)
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
x=print(func5(25))
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
x=print(func8([1,2,3,4,5]))
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr
x=print(func12([1, 2, 3, 4, 5], 1))
```

Programs on 1-d arrays like - sum of elements of array, searching an element inarray, finding minimum and maximum element in array, count the number ofeven and odd numbers in array. For all such programs, also find the timeComplexity, compare if there are multiple methods

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  sum=0
  for i in arr:
    sum=sum+i
  return sum
```

```python
#call the function
#arr = [12, 3, 4, 15,45,78,450]
#print(sum_arr)
#this funtion returning the sum of all element of array
import time
start = time.time()
arr = [12, 3, 4, 15,45,78,450]
print(sum_arr(arr))
end=time.time()
print(end-start)
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
  max=arr[0]
  min=arr[0]
  for i in arr:
    if i>max:
      max=i
    elif i<min:
      min=i
    return max,min
```

```python
# call sum_arr to find sum of elements in given array for arr = [12, 1234, 45, 67, 1]
import time
start = time.time()
arr = [12, 1234, 45, 67, 1]
print(sum_arr(arr))
end=time.time()
print(end-start)

```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def counting_Evenodd(arr):
  even=0
  odd=0
  for i in arr:
    if 1%2==0:
      even+=1
    else:
      odd+=1
  return even,odd

```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr,target):
  for i in range(len(arr)):
    if arr[i]==target:
      return i
  return -1
```

```python
# call the function liner_Search for array = [1, 3, 5, 7, 9]
import time
start = time.time()
array = [1, 3, 5, 7, 9]
x=print( linear_Search(arr,5))
end=time.time()
print(end-start)

```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr,target):
  low=0
  high=len(arr)-1
  while low<=high:
    mid=(low+high)//2
    if arr[mid]==target:
      return mid
    elif arr[mid]<target:
      low=mid+1
    else:
      high=mid-1
  return -1
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
import time
start = time.time()
arr = [1, 3, 5, 7, 9,15,16,14,45]
x=print( Binary_Search(arr,5))
end=time.time()
print(end-start)
```
