# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):
    dist = {}
    visited = set()
    parent = {}

    for node in graph:
        dist[node] = float('inf')
        parent[node] = None

    dist[start] = 0

    while len(visited) < len(graph):
        current = None

        for node in graph:
            if node not in visited:
                if current is None or dist[node] < dist[current]:
                    current = node

        if current is None:
            break

        visited.add(current)

        for neighbor in graph[current]:
            weight = graph[current][neighbor]

            if dist[current] + weight < dist[neighbor]:
                dist[neighbor] = dist[current] + weight
                parent[neighbor] = current

    return dist, parent
```

```python
graph = {}

nodes = int(input("How many nodes? "))

for _ in range(nodes):
    n = input("Enter node name: ").strip()
    graph[n] = {}

edges = int(input("How many edges? "))

for _ in range(edges):
    u = input("From: ").strip()
    v = input("To: ").strip()
    w = int(input("Weight: "))

    if w < 0:
        print("Negative weights not allowed!")
        continue

    if u not in graph or v not in graph:
        print("Invalid nodes!")
        continue

    graph[u][v] = w
    graph[v][u] = w
```
